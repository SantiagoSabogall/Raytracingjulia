## Funciones

```julia
function f(x)
    a = x^2
    return println("esta es mi primera funcion", a)
end
f(pi)
```

```julia
function signo(x)
    if x > 0
        return "positivo"
    elseif x < 0
        return "negativo"
    else
        return "cero"
    end
    
end

signo(5)
```

```julia
function escala(x, factor)
    return x * factor
end

println(escala(10, 2))
println(escala(10, 0.5))
```

```julia
function afina(x, base, factor)
    return base  + factor * x
end

afina(5, 10, 2)
```

```julia
function stats_basics(a,b,c)
    mean = (a + b + c) / 3
    median = sort([a, b, c])[2]
    range = maximum([a, b, c]) - minimum([a, b, c])
    return println("mean: $mean, median: $median, range: $range")
    
end
stats_basics(1, 2, 3)
```

```julia
mutable struct Structures
    blackhole
    in_edge::Float64
    out_edge::Float64
    corotating::Bool
end

function Structures(blackhole; R_min::Union{Nothing,Real}=nothing,
                              R_max::Real=20.0,
                              corotating::Bool=true)

    # Aquí replicamos tus ifs de Python
    local in_edge
    if isnothing(R_min)
        if corotating
            in_edge = blackhole.ISCOco
        else
            in_edge = blackhole.ISCOcounter
        end
    else
        in_edge = float(R_min)
    end

    Structure(blackhole, in_edge, float(R_max), corotating)
end

```

```julia
using Interpolations

# Datos conocidos
x = 0:10
y = @. sin(x)

# Crear una interpolación lineal
itp = interpolate(y, BSpline(Cubic(Line(OnGrid()))))

# Convertir a una función continua
itp_f = Interpolations.scale(itp, x)

# Evaluar en un punto intermedio
itp_f(4.5)   # ≈ sin(4.5)
```

```julia
using Plots

plot(x,itp)
scatter!(x, y)
scatter!(x, itp_f)
```
